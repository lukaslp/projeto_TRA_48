```python
import pandas as pd
from matplotlib import pyplot as plt


df_jan = pd.read_csv('../Dataset3_OntimePerformance_USA/Dataset3_OntimePerformance_USA/2019_01.csv', sep=',')
df_fev = pd.read_csv('../Dataset3_OntimePerformance_USA/Dataset3_OntimePerformance_USA/2019_02.csv', sep=',')
df_mar = pd.read_csv('../Dataset3_OntimePerformance_USA/Dataset3_OntimePerformance_USA/2019_03.csv', sep=',')
df_abr = pd.read_csv('../Dataset3_OntimePerformance_USA/Dataset3_OntimePerformance_USA/2019_04.csv', sep=',')
df_mai = pd.read_csv('../Dataset3_OntimePerformance_USA/Dataset3_OntimePerformance_USA/2019_05.csv', sep=',')
df_jun = pd.read_csv('../Dataset3_OntimePerformance_USA/Dataset3_OntimePerformance_USA/2019_06.csv', sep=',')
df_jul = pd.read_csv('../Dataset3_OntimePerformance_USA/Dataset3_OntimePerformance_USA/2019_07.csv', sep=',')
df_ago = pd.read_csv('../Dataset3_OntimePerformance_USA/Dataset3_OntimePerformance_USA/2019_08.csv', sep=',')
df_set = pd.read_csv('../Dataset3_OntimePerformance_USA/Dataset3_OntimePerformance_USA/2019_09.csv', sep=',')
df_out = pd.read_csv('../Dataset3_OntimePerformance_USA/Dataset3_OntimePerformance_USA/2019_10.csv', sep=',')
df_nov = pd.read_csv('../Dataset3_OntimePerformance_USA/Dataset3_OntimePerformance_USA/2019_11.csv', sep=',')
df_dez = pd.read_csv('../Dataset3_OntimePerformance_USA/Dataset3_OntimePerformance_USA/2019_12.csv', sep=',')

df_dez.head(10)
```

```python
df_main = pd.concat([df_jan,df_fev,df_mar,df_abr,df_mai,df_jun,df_jul,df_ago,df_set,df_out,df_nov,df_dez],ignore_index=True)
df_main
```

```python
## Apagando Voos cancelados

df_main = df_main[df_main.CANCELLED==0]
```

```python
df_main.drop(['YEAR','FL_DATE','TAXI_OUT','TAXI_IN','ARR_TIME','ACTUAL_ELAPSED_TIME','AIR_TIME','Unnamed: 26','CANCELLED','WHEELS_OFF','WHEELS_ON','DEP_TIME','ARR_DELAY'],inplace=True,axis=1)
```

```python
df_main.reset_index(inplace=True, drop=True)
```

```python
## Transformando os tempos em decimais, antes do ponto é hora, depois é minuto
df_main['CRS_DEP_TIME']=df_main['CRS_DEP_TIME']/100
df_main['CRS_ARR_TIME']=df_main['CRS_ARR_TIME']/100
```

```python
import seaborn as sns

df_main_barplot = df_main[['MONTH','DEP_DELAY']]

nobs = df_main_barplot.groupby(['MONTH']).count().reset_index()
nobs.head()
```

```python
plt.ylim(-50,+50)
boxplot_1T = sns.boxplot(x="MONTH", y="DEP_DELAY", data=df_main[df_main_barplot.MONTH.isin([1,2,3,4])])
medians = df_main[df_main_barplot.MONTH.isin([1,2,3,4])].groupby(['MONTH'])['DEP_DELAY'].median().values
nobs = df_main[df_main_barplot.MONTH.isin([1,2,3,4])].groupby(['MONTH'])['MONTH'].count().values
nobs = [str(x) for x in nobs.tolist()]
nobs = [i for i in nobs]
pos = range(len(nobs))
for tick,label in zip(pos,boxplot_1T.get_xticklabels()):
    boxplot_1T.text(pos[tick], medians[tick] + 3, nobs[tick],
    horizontalalignment='center', size='small', color='black', weight='bold')


```

```python
plt.ylim(-50,+50)
boxplot_2T = sns.boxplot(x="MONTH", y="DEP_DELAY", data=df_main[df_main_barplot.MONTH.isin([5,6,7,8])])
medians = df_main[df_main_barplot.MONTH.isin([5,6,7,8])].groupby(['MONTH'])['DEP_DELAY'].median().values
nobs = df_main[df_main_barplot.MONTH.isin([5,6,7,8])].groupby(['MONTH'])['MONTH'].count().values
nobs = [str(x) for x in nobs.tolist()]
nobs = [i for i in nobs]
pos = range(len(nobs))
for tick,label in zip(pos,boxplot_1T.get_xticklabels()):
    boxplot_2T.text(pos[tick], medians[tick] + 3, nobs[tick],
    horizontalalignment='center', size='small', color='black', weight='bold')

```

```python
plt.ylim(-50,+50)
boxplot_3T = sns.boxplot(x="MONTH", y="DEP_DELAY", data=df_main[df_main_barplot.MONTH.isin([9,10,11,12])])
medians = df_main[df_main_barplot.MONTH.isin([9,10,11,12])].groupby(['MONTH'])['DEP_DELAY'].median().values
nobs = df_main[df_main_barplot.MONTH.isin([9,10,11,12])].groupby(['MONTH'])['MONTH'].count().values
nobs = [str(x) for x in nobs.tolist()]
nobs = [i for i in nobs]
pos = range(len(nobs))
for tick,label in zip(pos,boxplot_1T.get_xticklabels()):
    boxplot_3T.text(pos[tick], medians[tick] + 10, nobs[tick],
    horizontalalignment='center', size='small', color='black', weight='bold')

```

```python
df_main['CRS_DEP_TIME_MM']=df_main.CRS_DEP_TIME.astype(str).str.extract('\.(.*)', expand=False).astype(int)
df_main['CRS_DEP_TIME_HH']=df_main.CRS_DEP_TIME.astype(str).str.extract('^([^.]*).*', expand=False).astype(int)
df_main['CRS_ARR_TIME_MM']=df_main.CRS_ARR_TIME.astype(str).str.extract('\.(.*)', expand=False).astype(int)
df_main['CRS_ARR_TIME_HH']=df_main.CRS_ARR_TIME.astype(str).str.extract('^([^.]*).*', expand=False).astype(int)
```
